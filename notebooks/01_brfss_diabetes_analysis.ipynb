```python
# Keep only valid target values for binary classification
df_clean = df_clean[df_clean["Diabetes_012"].isin([1.0, 3.0, 4.0])]

# Map target values:
# 1.0 = Diabetes
# 3.0 = No Diabetes
# 4.0 = Pre-diabetes / borderline diabetes, treated as No Diabetes
df_clean["Diabetes_012"] = df_clean["Diabetes_012"].replace({
    1.0: 1,
    3.0: 0,
    4.0: 0
})

print("Shape after cleaning target:", df_clean.shape)
print(df_clean["Diabetes_012"].value_counts())
```

```python
print("BMI missing before cleaning:")
print(df_clean["BMI"].isnull().sum())

print("\nBMI summary before scaling:")
print(df_clean["BMI"].describe())
```

```python
# BRFSS stores BMI as BMI * 100, so we divide by 100
df_clean["BMI"] = df_clean["BMI"] / 100

print("BMI missing after scaling:")
print(df_clean["BMI"].isnull().sum())

print("\nBMI summary after scaling:")
print(df_clean["BMI"].describe())
```

## Data Cleaning Summary

In this step, I cleaned the selected BRFSS 2024 dataset columns to prepare them for analysis and machine learning.

The target column `Diabetes_012` was converted into a binary classification target:
- `1` = Diabetes
- `0` = No Diabetes

Rows with unclear diabetes responses such as "Don't know", "Refused", missing values, and diabetes only during pregnancy were removed.

The `BMI` column was divided by 100 because BRFSS stores BMI values as BMI × 100.

For `PhysicalHealthDays` and `MentalHealthDays`:
- `88` was converted to `0`, meaning no unhealthy days.
- `77` and `99` were converted to missing values because they represent unknown or refused answers.

For binary health-related columns such as `Stroke`, `HeartDisease`, `PhysicalActivity`, `GeneralHealth`, and `HeavyAlcohol`, valid answers were mapped to `0` and `1`, while unknown or refused values were converted to missing values.

For `Smoker`, current smokers and former smokers were grouped as `1`, while never smokers were mapped to `0`.

For `Sex`, male was mapped to `1` and female was mapped to `0`.

For `Education` and `Income`, unknown values coded as `9` were converted to missing values.

Missing values were not dropped at this stage because they will be handled later using a preprocessing pipeline with imputation.
